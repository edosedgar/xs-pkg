```matlab
% ========================================================================
COVER_DIR = '/home/maxwell/Downloads/ad_cover/';
STEGO_DIR = '/home/maxwell/Downloads/ad_stego/';
% ========================================================================
files = dir(strcat(COVER_DIR, '*.jpg'));
pic_num = 1000;
for k=1:pic_num
    filename = strcat(COVER_DIR, files(k).name);
    [filepath, name, ext] = fileparts(filename);
    dct_coeff = jpeg_read(filename);
    [xpix, ypix] = size(dct_coeff.coef_arrays{1});
    stego_coeff = nsf5_simulation_cste_payload_color(dct_coeff , xpix * ypix * 8, 0.2, cputime);
    jpeg_write(stego_coeff, [STEGO_DIR name '.jpg']);
    disp(k);
end
```
